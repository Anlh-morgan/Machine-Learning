```python
import numpy as np
import pandas as pd
```

```python
def grad(x):
    return x**2 + 6*np.sin(x)

def cost(x):
    return 2*x + 6*np.cos(x)

def gradient_descent(x0, learning_rate, iterations):
    x = [x0]
    for i in range(iterations):
        x_new = x[-1] - learning_rate * cost(x[-1])
        if abs(grad(x_new)) < 1e-3:
            break
        x.append(x_new)
    return (x, i)

x0 = 0
learning_rate = 0.001
iterations = 1000000

results, num_iterations = gradient_descent(x0, learning_rate, iterations)

df = pd.DataFrame({'Iteration': range(len(results)),
                   'x Value': results})

df
```
